# Fase 2: Tratamento e Remoção de Duplicatas

Como realizamos buscas por múltiplos termos que possuem intersecção de conteúdo (ex: "adultização" e "15.211/2025"), é altamente provável que o algoritmo do YouTube tenha retornado o mesmo vídeo em buscas diferentes.

Para otimizar o consumo da nossa cota de API nas próximas fases, este notebook realiza uma etapa rápida de limpeza (Data Cleaning):
1. Carrega o dataset bruto de IDs.
2. Remove todas as linhas duplicadas baseadas na coluna `videoId`.
3. Exporta uma base limpa e consolidada.

**Output:** `unique_ids.csv`

```python
import pandas as pd
```

```python
# ids_lei_felca_1 tem uma coluna a menos (termo_busca) - não usar
df = pd.read_csv("./data/ids_lei_felca.csv")
display(df)
```

```python
df = df.drop_duplicates(subset=["videoId"])
print(len(df))
df.to_csv("unique_ids.csv", mode='w', index=False)
```
