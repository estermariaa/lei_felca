```python
pasta_transcricoes = "../transcriptions" 

if os.path.exists(pasta_transcricoes):
    arquivos_txt = [arquivo for arquivo in os.listdir(pasta_transcricoes) if arquivo.endswith('.txt')]
    
    arquivos_vazios = []    
    for arquivo in arquivos_txt:
        caminho_completo = os.path.join(pasta_transcricoes, arquivo)
        
        if os.path.getsize(caminho_completo) == 0:
            arquivos_vazios.append(arquivo)
        else:
            try:
                with open(caminho_completo, 'r', encoding='utf-8') as f:
                    conteudo = f.read().strip() 
                    if not conteudo:
                        arquivos_vazios.append(arquivo)
            except Exception as e:
                print(f"Erro ao ler o arquivo {arquivo}: {e}")
                
    print(f"Total de arquivos analisados: {len(arquivos_txt)}")
    print(f"Total de transcrições VAZIAS encontradas: {len(arquivos_vazios)}")
    
    if len(arquivos_vazios) > 0:
        print(f"\nExemplos de arquivos vazios: {arquivos_vazios[:5]}")

else:
    print("Pasta não encontrada.")
```
