Fase 15 - Adição do Atributo '```isShorts```' 

Nesta etapa, finalizamos o processo de distinção de vídeos curtos (Shorts) e vídeos longos.

O notebook processa as URLs redirecionadas para determinar a natureza de cada um dos registros, a partir da análise do formato da URL. As linhas com o padrão de URL ```/shorts/``` recebem o atributo isShorts com o valor 1, as linhas com o padrão de URL ```/watch?v=``` recebem o atributo isShorts com o valor 0.

````python
import pandas as pd
from urllib.parse import parse_qs, urlparse
````

````python
# Caminho (relativo ou absoluto) dos datasets de entrada
INPUT_DATASET_VIDEOS = ''
INPUT_DATASET_COMMENTS = ''

# Caminho (relativo ou absoluto) dos datasets de saída
OUTPUT_DATASET_VIDEOS = ''
OUTPUT_DATASET_COMMENTS = ''
````

````python
# Lê as URLs obtidas através da técnica de redirecionamento implementada nos passos anteriores
urls = list()
with open('../shorts/redirected.txt') as f:
    for line in f:
        urls.append(line)
````

````python
num_shorts = 0
num_videos = 0

# Mapeia cada um dos elementos se eles são shorts ou não
isShorts = dict()
for url in urls:
    parts = url.split('/')
    if 'shorts' in set(parts):
        id = str(parts[-1]).strip()
        isShorts[id] = 1
        print(id, url)
        num_shorts += 1
    else:
        # remove watch?
        parts = (parts[-1])[6:]

        params = parse_qs(parts) 

        id = params.get('v')[-1].strip()
        isShorts[id] = 0
        num_videos += 1
        
````

Adiciona a coluna 'isShorts' ao dataset de vídeos

````python
df = pd.read_csv(INPUT_DATASET_VIDEOS)
df['videoId'] = df['videoId'].astype(str)
df['isShorts'] = df['videoId'].map(isShorts).astype('Int64')

df.to_csv(OUTPUT_DATASET_VIDEOS, index=False)
````

Adiciona a coluna 'isShorts' ao dataset de comentários.

````python
df = pd.read_csv(INPUT_DATASET_COMMENTS)
df['videoId'] = df['videoId'].astype(str)
df['isShorts'] = df['videoId'].map(isShorts).astype('Int64')

df.to_csv(OUTPUT_DATASET_COMMENTS, index=False)
````

````python
print(f'Total de videos: {num_videos}')
print(f'Total de shorts: {num_shorts}')
print(f'Porcentagem de videos: {num_videos / (num_videos + num_shorts)}')
print(f'Porcentagem de shorts: {num_shorts / (num_videos + num_shorts)}')

print(f'Total de postagens (videos + shorts): {num_videos + num_shorts}')
````
