```python
import pandas as pd
```

```python
def insert_evaluation(df, index, evaluation):
    df.at[index, 'evaluation'] = evaluation
```

```python
df = pd.read_csv("../data/comments_to_evaluate.csv")
possible_eval = ["0", "1", "2", "s"]
```

```python
try:
    for index, row in df.iterrows():
        if pd.notna(row['evaluation']) and row['evaluation'] != "":
            continue
            
        print(f"Comentário {index + 1}")
        print(row['commentText'])
        
        evaluation = input("Avaliação [Neg: 0, Pos: 1, Neu: 2] ('s' salvar e sair): ")
        
        if evaluation in ['0', '1', '2']:
            insert_evaluation(df, index, evaluation)

        if evaluation.lower() == 's':
            break
            
        print()

except KeyboardInterrupt:
    print("\nInterrompido.")

df.to_csv("../data/comments_to_evaluate.csv", index=False)
print("Arquivo atualizado com sucesso!")
```

```python
import zipfile
import os

# Coloque o nome do seu zip e a pasta onde quer soltar os arquivos
caminho_zip = "../transcriptions.zip"
pasta_destino = "transcriptions"

# Cria a pasta de destino se ela não existir
os.makedirs(pasta_destino, exist_ok=True)

print("Extraindo milhares de arquivos... Aguarde!")

with zipfile.ZipFile(caminho_zip, 'r') as zip_ref:
    zip_ref.extractall(pasta_destino)

print("Extração concluída com sucesso!")
```
