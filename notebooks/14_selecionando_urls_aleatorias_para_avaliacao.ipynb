# Fase 14 - Avaliação dos Resultados (Shorts vs Vídeos Longos)

Nesta etapa, realizamos a validação manual da lógica de classificação implementada na fase anterior. O objetivo é garantir a integridade da estratégia de redirecionamento, confirmando se a separação entre vídeos curtos e longos foi realizada com precisão.

## Protocolo de Avaliação

Os avaliadores devem verificar o formato da URL corresponde à natureza do conteúdo:
- URLs https://www.youtube.com/shorts/{videoId}: Devem exibir obrigatoriamente a interface de Shorts do YouTube;
- URLs https://www.youtube.com/watch?v={videoId}: Devem exibir a interface padrão de vídeos longos.

Para garantir que dois avaliadores não analisem o mesmo material, o notebook gerencia dois arquivos principais:
1. Entrada ('../shorts/classify.txt'): contém a lista de URLs que aguardam avaliação.
2. Saída ('../shorts/evaluated.txt'): armazena as URLs já validadas até agora.

```python
from random import randint
```

```python
# Lê as urls obtidas atraves das requisições 
urls = list()
with open('../shorts/redirected.txt', mode='r') as f:
    for line in f: urls.append(line)
```

```python
# Lê as urls já avaliadas
evaluated = list()
with open('./output/evaluated.txt', mode='r') as f:
    for line in f: evaluated.append(line)
```

```python
# Define o número de urls a serem selecionadas para classificação
N_SELECT = 100 
```

```python
selected = list()
evaluated = set(evaluated)

for i in range(N_SELECT):
    j = randint(0, len(urls) - 1)

    while urls[j] in evaluated:
        j = randint(0, len(urls) - 1)

    evaluated.add(urls[j])
    selected.append(urls[j])

with open('..shorts/classify.txt', mode='w') as f:
    for line in selected: f.write(line) 

with open('..shorts/evaluated.txt', mode='a') as f:
    for line in selected: f.write(line)
```
