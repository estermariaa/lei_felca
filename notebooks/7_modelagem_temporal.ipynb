```python
import pandas as pd
from bertopic import BERTopic
```

```python
df = pd.read_csv("../data/preprocessed_title_description.csv")

# Transformando as datas para um padrão
df_model = df.dropna(subset=["clean_title_description", 'publishedAt']).copy()
df_model['data_limpa'] = pd.to_datetime(df_model['publishedAt']).dt.date.astype(str)
```

```python
docs = df_model["clean_title_description"].tolist()
timestamps = df_model['data_limpa'].tolist()
```

```python
print("Treinando o modelo base do BERTopic")
topic_model = BERTopic(verbose=True)

topics, probs = topic_model.fit_transform(docs)

print("Calculando a evolução temporal")
topics_over_time = topic_model.topics_over_time(docs, 
                                                timestamps, 
                                                nr_bins=20)

fig = topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=10)

fig.show()
```

```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

# 1. Configurando o agrupador para pegar expressões de até 3 palavras
# (ex: "projeto de lei", "liberdade de expressao")
vectorizer_model = CountVectorizer(ngram_range=(1, 3))

# 2. Construindo o Megazord com os parâmetros ajustados
topic_model = BERTopic(
    language="multilingual",
    min_topic_size=20,          # Um tópico precisa ter no mínimo 20 vídeos
    nr_topics="auto",           # Pede pro modelo fundir tópicos muito parecidos
    vectorizer_model=vectorizer_model,
    verbose=True
)

# 3. Treina o modelo normalmente
topics, probs = topic_model.fit_transform(docs)
print("Calculando a evolução temporal")
topics_over_time = topic_model.topics_over_time(docs, 
                                                timestamps, 
                                                nr_bins=20)

fig = topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=10)

fig.show()
```
