# Fase 11: Seleção de comentários para avaliação

Esse notebook seleciona uma amostra de 300 comentários por mês do período analisado para avaliação manual.

```python
import pandas as pd
```

```python
def select_sample(group):
    n = min(len(group), 300)
    return group.sample(n=n, random_state=42)
```

```python
df = pd.read_csv("../../comments_roberta.csv")

df['commentPublishedAt'] = pd.to_datetime(df['commentPublishedAt'])
df['year'] = df['commentPublishedAt'].dt.year
df['month'] = df['commentPublishedAt'].dt.month
df = df[df['language'] == 'pt']


df_sample = df.groupby(['year', 'month'], group_keys=False).apply(select_sample, include_groups=False)
df_sample = df_sample.reset_index(drop= True)
df_sample['evaluation'] = ""
print(df_sample.head())
df_sample.to_csv("comments_to_evaluate.csv", index=False)
```
