```python
import pandas as pd
```

```python
df = pd.read_csv("complete_dataset.csv")
df_comments = pd.read_csv("comments_dataset.csv")
display(df)
display(df_comments)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by="commentCount").reset_index(drop=True)

plt.figure()
plt.scatter(df_sorted.index, df_sorted["commentCount"])
plt.xlabel("Número de Vídeos (ordenados)")
plt.ylabel("Número de Comentários")
plt.title("Comentários por Vídeo (scatter)")
plt.show()
```

```python
plt.figure()
plt.scatter(df_sorted.index, df_sorted["commentCount"], s=10)

plt.ylim(0, 5000) 

plt.xlabel("Número de Vídeos (ordenados)")
plt.ylabel("Número de Comentários")
plt.title("Comentários por Vídeo (até 50k)")
plt.show()
```
